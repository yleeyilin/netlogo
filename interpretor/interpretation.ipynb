```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
courtyard_prediction = pd.read_csv('/Users/leeyilin/Downloads/netlogo/data/model/courtyard.csv')
utown_prediction = pd.read_csv('/Users/leeyilin/Downloads/netlogo/data/model/utown.csv')
```

```python
# plot graph 
x1 = courtyard_prediction['Category']
y1 = courtyard_prediction['air temp']

x2 = utown_prediction['Category']
y2 = utown_prediction['air temp']

plt.plot(x1, y1, label = "Courtyard Prediction of Air Temperature")
plt.plot(x2, y2, label = "UTown Green Prediction of Air Temperature")

plt.xlabel('Time')
plt.ylabel('Air Temperature')
plt.title('Air Temperature of Courtyard vs Utown Green')
plt.legend()
plt.show()
```

```python
# plot graph 
x1 = courtyard_prediction['Category']
y1 = courtyard_prediction['ground temp']

x2 = utown_prediction['Category']
y2 = utown_prediction['ground temp']

plt.plot(x1, y1, label = "Courtyard Prediction of Ground Temperature")
plt.plot(x2, y2, label = "UTown Green Prediction of Ground Temperature")

plt.xlabel('Time')
plt.ylabel('Ground Temperature')
plt.title('Ground Temperature of Courtyard vs Utown Green')
plt.legend()
plt.show()
```
