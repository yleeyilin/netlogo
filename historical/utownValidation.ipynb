```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
# load data
utown_observed = pd.read_csv('/Users/leeyilin/Downloads/netlogo/data/training test/utownrecorded.csv')
utown_predicted_sample = pd.read_csv('/Users/leeyilin/Downloads/netlogo/data/model/utown.csv').sample(n=8)
utown_predicted = pd.read_csv('/Users/leeyilin/Downloads/netlogo/data/model/utown.csv')
```

```python
# view observed courtyard data
utown_observed
```

```python
# view predicted courtyard data
utown_predicted.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=utown_predicted, x='Category', y='air temp', label='Scatterplot (Observation)')
sns.scatterplot(data=utown_observed, x='Category', y='air temp', label='Scatterplot (NetLogo)')
sns.regplot(data=utown_predicted, x= 'Category', y = 'air temp', scatter=False, color='red', label='Linear Regression (NetLogo)')
sns.regplot(data=utown_observed, x='Category', y='air temp', scatter=False, color='blue', label='Linear Regression (Observation)')
plt.grid(True)

plt.xlabel('Time')
plt.ylabel('Air Temperature')
plt.title('UTown Green Air Temperature in NetLogo vs Observed')
plt.legend()

plt.show()
observed = utown_observed['air temp']
predicted = utown_predicted_sample['air temp']

# Calculate R-squared value
r_squared = r2_score(observed, predicted)
print("R Square Value: " + str(r_squared))
```

```python
# Plot the filtered data
plt.figure(figsize=(12, 8))
sns.scatterplot(data=utown_predicted, x='Category', y='ground temp', label='Scatterplot (NetLogo)')
sns.scatterplot(data=utown_observed, x='Category', y='ground temp', label='Scatterplot (Observation)')
sns.regplot(data=utown_predicted, x='Category', y='ground temp', scatter=False, color='red', label='Linear Regression (NetLogo)')
sns.regplot(data=utown_observed, x='Category', y='ground temp', scatter=False, color='blue', label='Linear Regression (Observation)')
plt.grid(True)

plt.xlabel('Time')
plt.ylabel('Ground Temperature')
plt.title('UTown Green Ground Temperature in NetLogo vs Observed')
plt.legend()

plt.show()

observed = utown_observed['ground temp']
predicted = utown_predicted_sample['ground temp']

# Calculate R-squared value
r_squared = r2_score(observed, predicted)
print("R Square Value: " + str(r_squared))
```
