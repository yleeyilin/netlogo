```python
import pandas as pd
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
courtyard_model_df = pd.read_csv('courtyard.csv')
utown_model_df = pd.read_csv('utown.csv')
courtyard_recorded_df = pd.read_csv('utownrecorded.csv')
courtyard_recorded_df.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(courtyard_model_df, x = 'Category', y = 'air temp')
sns.regplot(data=courtyard_model_df, x='Category', y='air temp', scatter=False, color='red')
sns.regplot(data=courtyard_recorded_df, x='Category', y="Air temp", scatter=False, color='blue')
plt.grid(True)
```
